Nesta atividade, você deverá estruturar um pacote contendo funções para manipulação de matrizes. As matrizes devem ser tratadas como listas de listas.

Crie um módulo (matrix.py) que contenha as seguintes funções:
1. transpor_matriz(matriz)

* Regra: Deve transformar as linhas da matriz original em colunas e vice-versa.
* Entrada: Uma lista de listas $M \times N$.
* Saída: Uma nova lista de listas $N \times M$.

2. multiplicar_matriz(matriz_a, matriz_b)

* Regra: Deve realizar a multiplicação matricial clássica (Linha por Coluna).
* Validação: A função deve verificar se o número de colunas de A é igual ao número de linhas de B. Caso contrário, retorne uma mensagem de erro ou None.
* Entrada: Duas matrizes (listas de listas).
* Saída: Uma matriz resultante da multiplicação.

## matrix.py
Módulo com as funções de manipulação de matrizes.

````python
# matrix.py
# Para usar como módulo externo, salve este conteúdo em um arquivo chamado matrix.py

def transpor_matriz(matriz):
    """
    Transpõe uma matriz M x N, retornando uma nova matriz N x M.
    As linhas da matriz original tornam-se colunas e vice-versa.

    Parâmetros:
        matriz (list[list]): Matriz de entrada M x N.

    Retorna:
        list[list]: Matriz transposta N x M.
    """
    if not matriz or not matriz[0]:
        return []

    linhas = len(matriz)
    colunas = len(matriz[0])

    transposta = []
    for j in range(colunas):
        nova_linha = []
        for i in range(linhas):
            nova_linha.append(matriz[i][j])
        transposta.append(nova_linha)

    return transposta


def multiplicar_matriz(matriz_a, matriz_b):
    """
    Multiplica duas matrizes usando o método clássico (linha por coluna).
    Valida se o número de colunas de A é igual ao número de linhas de B.

    Parâmetros:
        matriz_a (list[list]): Primeira matriz (M x N).
        matriz_b (list[list]): Segunda matriz (N x P).

    Retorna:
        list[list]: Matriz resultante (M x P), ou None se dimensões incompatíveis.
    """
    linhas_a = len(matriz_a)
    colunas_a = len(matriz_a[0])
    linhas_b = len(matriz_b)
    colunas_b = len(matriz_b[0])

    if colunas_a != linhas_b:
        print("Erro: o número de colunas de A deve ser igual ao número de linhas de B.")
        print(f"  A é {linhas_a}x{colunas_a}, B é {linhas_b}x{colunas_b} — incompatível.")
        return None

    # Inicializa a matriz resultado com zeros
    resultado = [[0] * colunas_b for _ in range(linhas_a)]

    for i in range(linhas_a):
        for j in range(colunas_b):
            for k in range(colunas_a):
                resultado[i][j] += matriz_a[i][k] * matriz_b[k][j]

    return resultado
````

## main.py
Arquivo principal que utiliza o módulo `matrix.py`.

````python
# main.py
# Este bloco simula o main.py que importaria o matrix.py.
# Como estamos no mesmo notebook, as funções já estão definidas acima.

def exibir_matriz(matriz):
    """Exibe uma matriz formatada linha por linha."""
    for linha in matriz:
        print(linha)


# -----------------------------------------------
# Exemplo 1: Transposição
# -----------------------------------------------
print("=== Transposição de Matriz ===")
A = [[1, 2], [3, 4], [5, 6]]

print("Matriz original (3x2):")
exibir_matriz(A)

transposta = transpor_matriz(A)
print("\nMatriz transposta (2x3):")
exibir_matriz(transposta)
print(f"\nResultado: {transposta}")


# -----------------------------------------------
# Exemplo 2: Multiplicação
# -----------------------------------------------
print("\n=== Multiplicação de Matrizes ===")
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

print("Matriz A (2x2):")
exibir_matriz(A)

print("\nMatriz B (2x2):")
exibir_matriz(B)

produto = multiplicar_matriz(A, B)
if produto is not None:
    print("\nMatriz A x B (2x2):")
    exibir_matriz(produto)
    print(f"\nResultado: {produto}")


# -----------------------------------------------
# Exemplo 3: Multiplicação com dimensões incompatíveis
# -----------------------------------------------
print("\n=== Teste de Validação (dimensões incompatíveis) ===")
C = [[1, 2, 3]]
D = [[4, 5], [6, 7]]
resultado_invalido = multiplicar_matriz(C, D)
````
